```python
df=pd.DataFrame(lusitania_df, columns=['Age'])
```

```python
df['months age']=df['Age'].str.replace('(\d\d-......)', '1')
df['real age'] = df['months age'].str.extract('(\d\d?)', expand=True)
```

```python
df.drop(['Age', 'months age'], inplace = True, axis=1)
df.head()
```

```python
df=df.convert_objects(convert_numeric=True)

df=df.dropna()
df
```

#### What does the group-wise survival rate imply about circumstances during these two accidents?

```python
lusitania_long= pd.melt(lusitania_df, id_vars=['Family name'])
lusitania_long.sort_values('Family name', axis=0)


```

```python
def first_item(series):
    item = series.iloc[0]
    return item

pivot_lusitania= pd.pivot_table(lusitania_long, index='Family name', values='value',
                                columns=['variable'], aggfunc=first_item)
pivot_lusitania.reset_index()
```

```python
titanic_long= pd.melt(titanic_df, id_vars=['name'])
titanic_long.sort_values('name', axis=0)
```

```python
def first_item(series):
    item = series.iloc[0]
    return item

pivot_titanic= pd.pivot_table(titanic_long, index='name', values='value',
                                columns=['variable'], aggfunc=first_item)
pivot_titanic.reset_index()
```

For Titanic, it was more likely that you survived if you were a female that was in a higher class. It was also more likely that you survived if you were able to be on one of the lifeboats.

For the Lusitania, it was more likely that you survived if you were a passenger and not a crew member. It was also more likely that you survived if you were a female as well.
